## 1: 2D Line Plot

2D plot is apply for:

Bivariate analyses -> using two columns on x , y axis

catagorical->numarical and numarical->numarical

Use case - Time Series data

```python
# plotting simple plot
price = [48000,54000,57000,49000,47000,45000] # numarical column
year = [2015,2016,2017,2018,2019,2020]  # catagorical column

# plt.plot(x-axis, y-axis)
plt.plot(year,price)
```

```python
# limit the axis
price = [48000,54000,57000,49000,47000,45000,4500000] 
year = [2015,2016,2017,2018,2019,2020,2021]  

plt.plot(year,price)
```

```python
# limit the axis
price = [48000,54000,57000,49000,47000,45000,4500000] 
year = [2015,2016,2017,2018,2019,2020,2021]  

plt.plot(year,price)
plt.ylim(0,75000)  # we want to see the price betwwen 0 to 75000
```

```python
# limit the axis
price = [48000,54000,57000,49000,47000,45000] 
year = [2015,2016,2017,2018,2019,2020]  

plt.plot(year,price)
plt.xlim(2018,2020) # we want to see the yeeat only between (2018,2020)
```

## 3: Bar Chart

Bar Chart is apply for:

Bivariate analyses -> using two columns on x , y axis

numarical vs catagorical

Use case - Aggregate analysis of groups

```python
# simple bar chart
# plt.bar(x-axis,y-axis)
children = [10,20,40,10,30]
color = ["Red", "Blue", "Green", "Yellow", "Pink"]

plt.bar(color,children)
```

```python
# color
plt.bar(color,children,color="green")
```

```python
# horizantal barchart
#  barh()
plt.barh(color,children)
```

```python
# multiple bar chart
df = pd.read_csv("batsman_session_record.csv")
```

```python
df
```

```python
# plot bar chart of each batsman for 2015 session
plt.bar(df["batsman"],df["2015"])

plt.title("comparison of 5 top batsmain in 2015 session")
```

```python
# plot multiples bar graph 
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2)

```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2)

plt.xticks(np.arange(df.shape[0]),df["batsman"])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2)

plt.xticks(np.arange(df.shape[0]),df["batsman"])

plt.show()
```

```python
# xticks -> used to rename
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2,color="red")
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2,color="green")
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2,color="blue")

plt.xticks(np.arange(df.shape[0]),df["batsman"])

plt.show()
```

```python
# problem

children = [10,20,40,10,30]
color = ["Red  Red Red", "Blue Blue Blue Blue ", "Green Green Green Green ", "Yellow Yellow Yellow Yellow ", "Pink Pink Pink Pink "]

plt.bar(color,children)
```

```python
# xticks()  rotation
children = [10,20,40,10,30]
color = ["Red  Red Red", "Blue Blue Blue Blue ", "Green Green Green Green ", "Yellow Yellow Yellow Yellow ", "Pink Pink Pink Pink "]

plt.bar(color,children)
plt.xticks(rotation="vertical")
plt.show()
```

```python
# stack bar chart
plt.bar(df["batsman"],df["2015"])
plt.bar(df["batsman"],df["2016"],bottom=df["2015"])
plt.bar(df["batsman"],df["2017"],bottom=(df["2015"] + df["2016"]))
```

```python
plt.bar(df["batsman"],df["2015"],label="2015")
plt.bar(df["batsman"],df["2016"],bottom=df["2015"],label="2016")
plt.bar(df["batsman"],df["2017"],bottom=(df["2015"] + df["2016"]),label="2017")

plt.legend()
plt.show()
```

## 5: Pie chart

Pie chart is apply for:

Univariate/Bivariate Analyses -> using  one or two columns

catagorical vs numarical

Use case - To find contribution on a standard scale  (percentile)

```python
# simple data
data = [23,45,100,20,49]

plt.pie(data)
plt.show()
```

```python
# label
data = [23,45,100,20,49]
subjects = ["english","physics","maths","urdu","pak study"]
plt.pie(data,labels=subjects)
plt.show()
```

```python
# from data set
df = pd.read_csv("match_score.csv")
```

```python
df
```

```python
plt.pie(df["batsmain_runs"], labels=df["batsman"])
plt.show()
```

```python
# show the percentage
plt.pie(df["batsmain_runs"], labels=df["batsman"],autopct="%0.1f%%")
plt.show()
```

```python
# show the percentage and color
plt.pie(df["batsmain_runs"], labels=df["batsman"],autopct="%0.1f%%",colors=["blue","green","yellow","pink","cyan","brown"])
plt.show()
```

```python
# shadow
plt.pie(df["batsmain_runs"], labels=df["batsman"],autopct="%0.1f%%",shadow=True)
plt.show()
```

```python
# explode 
# we want to get AB de Velliors	who is on 1st position in dataset
plt.pie(df["batsmain_runs"], labels=df["batsman"],autopct="%0.1f%%",explode=[ 0.1,0,0,0,0,0])
plt.show()
```
