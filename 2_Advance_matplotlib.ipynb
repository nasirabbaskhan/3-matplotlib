### Subplot

Subplot is using ploting multiples graph

```python
# ploting scatter graph
# plt.figure(figsize=(15,6)) #contro the size

plt.scatter(df["avg"],df["strike_rate"])
plt.title("something")
plt.xlabel("avg")
plt.ylabel("strike_rate")

plt.show()
```

```python
# A differ wat to ploting scatter graph --> 2nd method
# fig, ax = plt.subplots(figsize=(15,6)) # contro the size
fig, ax = plt.subplots()
ax.scatter(df["avg"],df["strike_rate"])
ax.set_title("something")
ax.set_xlabel("avg")
ax.set_ylabel("strike_rate")

fig.show()
```

```python
# plotting 2 graph side by side
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(df["avg"],df["strike_rate"],color="red")
ax[1].scatter(df["avg"],df["runs"])

ax[0].set_title("avg vs strike rare")
ax[0].set_ylabel("strike_rate")

ax[1].set_title("avg vs runs")
ax[1].set_xlabel("avg")
ax[1].set_ylabel("runs")

```

```python
# plotting 4 graph side by side
fig, ax = plt.subplots(nrows=2,ncols=2)
ax[0,0].scatter(df["avg"],df["strike_rate"],color="red")
ax[0,1].scatter(df["avg"],df["runs"])
ax[1,0].hist(df["avg"])
ax[1,1].hist(df["strike_rate"],color="green")

# ax[0].set_title("avg vs strike rare")
# ax[0].set_ylabel("strike_rate")

# ax[1].set_title("avg vs runs")
# ax[1].set_xlabel("avg")
# ax[1].set_ylabel("runs")
```

### 3D Surface Plot

```python
x= np.linspace(-10,10,100)
y= np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure()

ax = plt.subplot(projection="3d")

p=ax.plot_surface(xx,yy,z)
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure()

ax = plt.subplot(projection="3d")

p=ax.plot_surface(xx,yy,z)
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure()

ax = plt.subplot(projection="3d")

p=ax.plot_surface(xx,yy,z)
fig.colorbar(p)
```

### Contour Plot
it represent 3D graph into 2D

```python
z = xx**2 + yy**2
fig = plt.figure()

ax = plt.subplot()

p=ax.contour(xx,yy,z)
fig.colorbar(p)
```

```python
z = xx**2 + yy**2
fig = plt.figure()

ax = plt.subplot()

p=ax.contourf(xx,yy,z)
fig.colorbar(p)
```
